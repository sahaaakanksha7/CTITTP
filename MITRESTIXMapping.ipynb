```python
import requests

def get_stix_data(url):
    """
    Fetches STIX data for a given MITRE ATT&CK version from a URL.

    Parameters
    ----------
    url : str
        URL to fetch the STIX data.

    Returns
    -------
    dict
        Parsed STIX data as JSON.
    """
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        raise ValueError(f"Error fetching data from {url}. Status code: {response.status_code}")

def track_ttps_across_versions(domain_versions):
    """
    Tracks all TTPs across all ATT&CK versions, including their introduction, deprecation, revocation, and domain.

    Parameters
    ----------
    domain_versions : list of tuples
        List of tuples containing (domain, url) pairs for the STIX data.

    Returns
    -------
    dict
        Dictionary containing TTP metadata.
    """
    ttp_map = {}

    for domain, url in domain_versions:
        # Extract version from the URL
        version = url.split("/")[-1].replace(f"{domain}-attack-", "").replace(".json", "")
        
        try:
            print(f"Fetching data for {domain} version {version}...")
            stix_data = get_stix_data(url)
            #print(stix_data)
        except ValueError as e:
            print(e)
            continue

        # Iterate over STIX objects to find attack patterns
        for obj in stix_data.get('objects', []):
            if obj.get('type') == 'attack-pattern':
                #print(obj)
                ttp_id = None

                for ref in obj.get('external_references', []):
                    if ref.get('source_name') in ['mitre-attack', 'mitre-ics-attack', 'mitre-attack-mobile']:
                        ttp_id = ref.get('external_id')
                        break

                if not ttp_id:
                    continue

                # If the TTP is already in the map, ensure the first appearance is the earliest one
                if ttp_id in ttp_map:
                    # Update first appearance if this version is earlier
                    if version < ttp_map[ttp_id]['first_appearance']:
                        ttp_map[ttp_id]['first_appearance'] = version
                else:
                    # Otherwise, add a new entry for the TTP
                    ttp_map[ttp_id] = {
                        'first_appearance': version,
                        'deprecated_version': None,
                        'revoked_version': None,
                        'domain': domain
                    }

                # Track deprecated version if not already set
                if obj.get('x_mitre_deprecated') and ttp_map[ttp_id]['deprecated_version'] is None:
                    ttp_map[ttp_id]['deprecated_version'] = version

                # Track revoked version, ensuring it captures the first instance
                if obj.get('revoked'):
                    if ttp_map[ttp_id]['revoked_version'] is None or version < ttp_map[ttp_id]['revoked_version']:
                        ttp_map[ttp_id]['revoked_version'] = version

    return ttp_map



# Combine all versions
all_versions = mobile_versions 
#+ mobile_versions + enterprise_versions


def extract_domain_from_url(url):
    """
    Extracts the domain (e.g., 'ics-attack', 'mobile-attack') from the URL.

    Parameters
    ----------
    url : str
        The URL of the STIX data file.

    Returns
    -------
    str
        The domain extracted from the URL.
    """
    # Extract the part between 'master/' and the next '/'
    return url.split("/")[-2]

def prepare_domain_versions(url_list):
    """
    Prepares a list of (domain, URL) tuples from the list of URLs.

    Parameters
    ----------
    url_list : list
        List of URLs.

    Returns
    -------
    list
        List of tuples where each tuple contains (domain, URL).
    """
    return [(extract_domain_from_url(url), url) for url in url_list]

# Prepare the domain versions as tuples (domain, URL)
domain_versions = prepare_domain_versions(all_versions)

# Track the TTPs across all domains and versions
ttp_data = track_ttps_across_versions(domain_versions)


# Output the results
for ttp_id, info in ttp_data.items():
    print(f"TTP ID: {ttp_id}")
    print(f"  Domain: {info['domain']}")
    print(f"  First Appearance: {info['first_appearance']}")
    print(f"  Deprecated Version: {info['deprecated_version']}")
    print(f"  Revoked Version: {info['revoked_version']}\n")
```
